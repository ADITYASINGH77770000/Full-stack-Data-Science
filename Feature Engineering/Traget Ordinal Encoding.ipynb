## Target Ordinal Encoding 

```python
import pandas as pd
df = pd.DataFrame({
    'city': ['New York', 'Los Angeles', 'Chicago', 'Houston', 'Phoenix'],
    'price': [200,150,300,180,320]
})
```

```python
df
```

```python
mean_price=df.groupby('city') ['price'].mean()
```

```python
mean_price
```

```python
df['city_encoded']=df['city'].map(mean_price)
```
