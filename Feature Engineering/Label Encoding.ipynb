## Label/Ordinal Encoding

```python
import pandas as pd
```

```python
## Create a sample dataframe
df=pd.DataFrame({
    'Color':['Red','Blue','Green','Blue','Red','Green','Red']
})  
```

```python
df.head()
```

```python
from sklearn.preprocessing import LabelEncoder
lbl_encoder=LabelEncoder()
```

```python
lbl_encoder.fit_transform(df['Color'])
```

```python
lbl_encoder.transform(['Green','Red','Blue'])
```

## Ordinal Encoding 

```python
from sklearn.preprocessing import OrdinalEncoder
```

```python
## create a dataframe with an ordinal feature
df=pd.DataFrame({
    'Size':['Small','Medium','Large','Medium','Small','Large','Large']
})
```

```python
df
```

```python
## crate an instance of OrdinalEncoder and then fit_transform
ord_encoder=OrdinalEncoder(categories=[['Small','Medium','Large']])
```

```python
ord_encoder.fit_transform(df[['Size']])
```

```python
ord_encoder.transform([['Medium'],['Large'],['Small']])
```
