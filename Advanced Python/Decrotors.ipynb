## Decorators
Decorators are a powerful and flexible feature in Python that allows you to modify the behavior of a function or class method. They are commonly used to add functionality to functions or methods without modifying their actual code. This lesson covers the basics of decorators, including how to create and use them.

```python
## function copy
def welcome():
    return "Welcome to Python Decorators"
welcome()
```

```python
wel=welcome
print(wel())
del welcome
print(wel())
```

```python
## Decrotors
def main_welcome(func):
    def sub_welcome_method():
        print("Wecome to the advance python course")
        func()
        print("Enjoy the course")
    return sub_welcome_method
```

```python
def course_introduction():
    print("This is a advance python course")
course_introduction()
```

```python
main_welcome(course_introduction)()
```

```python
@main_welcome
def course_introduction():
    print("This is a advance python course")
```

```python
## Decrotors

def my_decrotor(func):
    def wrapper():
        print("Before executing the function")
        func()
        print("After executing the function")
    return wrapper
```

```python
@my_decrotor
def say_hello():
    print("Hello, World!")
```

```python
say_hello()
```

```python
## Decrotors with arguments
def repeat(n):
    def decroator(func):
        def wrappe(*args,**kwargs):
            for _ in range(n):
                func(*args,**kwargs)
        return wrappe
    return decroator
```

```python
@repeat(3)
def say_hello():
    print("Hello, World!")
```

```python
say_hello()
```
